```python
import pandas as pd
sales = {
    'Day' : ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
    'Revenue' : [1200, 1500, 900, 2000, 1800]
}

data = pd.DataFrame(sales)
```

```python
print('Total Revenue \n', data['Revenue'].sum())
```

```python
print('Average daily Revenue \n', data['Revenue'].mean())
```

```python
print('Highest Revenue \n', data['Revenue'].max())
```

```python
print('Highest Revenue \n', data[data['Revenue'] > data['Revenue'].mean()])
```

```python
data.plot(x='Revenue', y='Day', kind='hist')
```

```python
pd.set_option("plotting.backend", "plotly")
data.plot(x='Revenue', y='Day', kind='hist')
```
